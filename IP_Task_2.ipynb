```python
# Tic-Tac-Toe with Minimax AI

# Constants for the game
EMPTY = ' '
PLAYER = 'X'
AI = 'O'

# Function to print the board
def print_board(board):
    print("-------------")
    for row in board:
        print(f"| {' | '.join(row)} |")
        print("-------------")

# Function to check if the board is full
def is_board_full(board):
    return all(cell != EMPTY for row in board for cell in row)

# Function to check if a player has won
def check_winner(board, player):
    # Check rows, columns, and diagonals
    for i in range(3):
        if all(board[i][j] == player for j in range(3)):  # Rows
            return True
        if all(board[j][i] == player for j in range(3)):  # Columns
            return True
    if all(board[i][i] == player for i in range(3)):  # Main diagonal
        return True
    if all(board[i][2 - i] == player for i in range(3)):  # Anti-diagonal
        return True
    return False

# Minimax algorithm
def minimax(board, depth, is_maximizing):
    # Base cases: check for terminal states
    if check_winner(board, AI):
        return 1
    if check_winner(board, PLAYER):
        return -1
    if is_board_full(board):
        return 0

    if is_maximizing:
        best_score = -float('inf')
        for i in range(3):
            for j in range(3):
                if board[i][j] == EMPTY:
                    board[i][j] = AI
                    score = minimax(board, depth + 1, False)
                    board[i][j] = EMPTY
                    best_score = max(score, best_score)
        return best_score
    else:
        best_score = float('inf')
        for i in range(3):
            for j in range(3):
                if board[i][j] == EMPTY:
                    board[i][j] = PLAYER
                    score = minimax(board, depth + 1, True)
                    board[i][j] = EMPTY
                    best_score = min(score, best_score)
        return best_score

# Function to find the best move for the AI
def find_best_move(board):
    best_score = -float('inf')
    best_move = (-1, -1)
    for i in range(3):
        for j in range(3):
            if board[i][j] == EMPTY:
                board[i][j] = AI
                score = minimax(board, 0, False)
                board[i][j] = EMPTY
                if score > best_score:
                    best_score = score
                    best_move = (i, j)
    return best_move

# Main game loop
def play_game():
    # Initialize the board
    board = [[EMPTY for _ in range(3)] for _ in range(3)]
    print("Welcome to Tic-Tac-Toe!")
    print_board(board)

    while True:
        # Player's turn
        player_row, player_col = -1, -1
        while True:
            try:
                player_row = int(input("Enter the row (0, 1, 2): "))
                player_col = int(input("Enter the column (0, 1, 2): "))
                if board[player_row][player_col] == EMPTY:
                    break
                else:
                    print("That cell is already occupied. Try again.")
            except (ValueError, IndexError):
                print("Invalid input. Please enter numbers between 0 and 2.")

        board[player_row][player_col] = PLAYER
        print_board(board)

        # Check if the player has won
        if check_winner(board, PLAYER):
            print("Congratulations! You won!")
            break

        # Check if the board is full
        if is_board_full(board):
            print("It's a tie!")
            break

        # AI's turn
        print("AI is making a move...")
        ai_row, ai_col = find_best_move(board)
        board[ai_row][ai_col] = AI
        print_board(board)

        # Check if the AI has won
        if check_winner(board, AI):
            print("AI wins! Better luck next time.")
            break

        # Check if the board is full
        if is_board_full(board):
            print("It's a tie!")
            break

# Start the game
if __name__ == "__main__":
    play_game()
```
